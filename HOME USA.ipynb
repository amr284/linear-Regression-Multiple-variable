# Home prices in monroe township , NJ(USA)

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# import csv file 
df = pd.read_csv("homeprices22.csv")
df
```

price = m1 * area + m2 * bedrooms + m3 * age +b

-----------------------------------------------------------------
price ---> dependant variable 

area , bedrooms , age ---> independant variable (features)

m1 , m2 , m3 ----> coefficients

b---> intercept

---------------------------------------------------------------------

```python
# Data Preprocessing: Fill NA values with median value of a column
median_bedrooms= df.bedrooms.median()
median_bedrooms
```

```python
# fill this column's na values  with value by fillna
df.bedrooms=df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
# create linear regression object so i have imported linear model 
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms' , 'age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old

```python
reg.predict([[3000,3,40]])
```

Find price of home with 2500 sqr ft area, 4 bedrooms, 5 year old

```python
reg.predict([[2500,4,5]])
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('price (us$)')
plt.scatter(df.area,df.price , color='red' , marker='+')

```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

# save model using  pickle

```python
# saving model using pickle into afile 
import pickle
with open (' model_pickle' , 'wb') as f :
    pickle.dump(model,f)
```
